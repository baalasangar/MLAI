```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
## QQ plot - Chek is the data is normal distributed
x = np.array([26, 33, 65, 28, 34, 55, 25, 44, 50, 36, 26, 37, 43, 62, 35, 38, 45, 32, 28, 34])
normal_dis = np.random.standard_normal(21)
normal_dis =  np.sort(normal_dis)
perc = np.linspace(0,100,21)
qun_x = np.percentile(x, perc)
sns.regplot(normal_dis,qun_x)
```

```python
## QQ plot - Chek is the data is uniform distributed
uniform_dis = np.random.uniform(low=0,high=1, size=21)
sns.regplot(uniform_dis,qun_x)
```
