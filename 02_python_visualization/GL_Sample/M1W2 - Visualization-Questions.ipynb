# Q.1 Import the libraries




```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="darkgrid")
%matplotlib inline
```

# Q.2 Get the Dataset

```python
df = pd.read_csv("tips.csv")
```

```python
df.dtypes
```

```python
df.head(10)
```

# Q3. Plot the distribution of total_bill in the given dataset.

```python
sns.distplot(df["total_bill"]);
```

# Q.4 - Plot the destribution plot for tips.

```python
sns.distplot(df["tip"]);
```

# Q5. Make a joint plot between tip and total bill.

```python
sns.jointplot(data=df,x="tip",y="total_bill")
```

# Q6. Give additional attribute kind = "hex" in the above plot.

```python
sns.jointplot(data=df,x="tip",y="total_bill",kind="hex")
```

# Q7. Make pairplot between total bill and tip.

```python
sns.pairplot(df[["tip","total_bill"]])
```

# Q.8 -  Make striplot between sex and tip.

```python
sns.catplot(data=df,x="sex",y="tip")
```

```python
sns.stripplot(data=df,x="sex",y="tip")
```

# Q.9 - What other plots you can make for tip and sex variable.Plot them.

```python
# Swarmplot 
sns.catplot(kind="swarm", data=df,x="sex",y="tip")
```

```python
sns.catplot(kind="box", data=df,x="sex",y="tip")
```

```python
sns.catplot(kind="bar", data=df,x="sex",y="tip")
```

```python
sns.catplot(kind="point", data=df,x="sex",y="tip")
```

# Q.10 Make bar plot with day and total bill.

```python
sns.catplot(kind="bar", data=df,x="day",y="total_bill")
```

# Q.11 - Make a plot to count total males and females for each day.

```python
sns.countplot(data=df, x="day", hue="sex")
```

# Q.12 -  Use factorplot to plot multiple categorical variables like day, smoker, peoplecount.

```python
sns.catplot(x = "day",
               y = "size",
               hue = "size",
               col = "smoker",
               data = df,
               kind = "bar"
              )
```

# Q.13 - Use lmplot() to plot the relation between total bill and tip.


```python
sns.lmplot(data=df, x="total_bill", y="tip")
```

# Q.14 - Plot the above graph for male and female.

```python
sns.lmplot(data=df, x="total_bill", y="tip", hue="sex")
```

# Q.14 - Plot the Q.13 graph for smoker and non-smoker.

```python
sns.lmplot(data=df, x="total_bill", y="tip", hue="smoker")
```
