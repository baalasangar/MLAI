```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
import cv2
```

```python
# Load data
data = pd.read_csv("./Cars Images/Car_tag_data.csv")
data.head()
```

```python
data.iloc[0]["IMAGE_NAME"]
```

```python
# Visualize image 

fig=plt.figure(figsize=(20,20))
columns=3
rows=5
   
fig=plt.figure(figsize=(12,12))
columns=3
rows=5
for i in range(0, columns*rows):
    #print(i)
    img = cv2.imread('./Cars Images/'+str(data.iloc[i]["IMAGE_NAME"]),cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    fig.add_subplot(rows,columns,i+1)
    plt.imshow(img)
    #plt.title(labels[i])
    plt.axis('off')
plt.show()    
```

```python
# Visualize image - with color and Make of the car

fig=plt.figure(figsize=(20,20))
columns=3
rows=5
   
fig=plt.figure(figsize=(12,12))
columns=3
rows=5
for i in range(0, columns*rows):
    #print(i)
    img = cv2.imread('./Cars Images/'+str(data.iloc[i]["IMAGE_NAME"]),cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    fig.add_subplot(rows,columns,i+1)
    plt.imshow(img)
    plt.title(data.iloc[i]["MAKE"] + "--"+ data.iloc[i]["COLOR"] )
    plt.axis('off')
plt.show()   
```
