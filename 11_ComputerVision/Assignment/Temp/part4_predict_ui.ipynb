```python
from tkinter import *
from tkinter import filedialog
import ntpath
from tensorflow.keras.models import load_model
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



```

```python

ROOT_PATH = str()

pretrained_model = load_model('./flower_prediction_model.h5')
pretrained_model.load_weights('./flower_prediction_weights.h5')
  
# Function for opening the
# file explorer window
def browseFiles(filenameStr):
    filename = filedialog.askopenfilename(initialdir = "/",
                                          title = "Select a File",
                                          filetypes = (("Text files",
                                                        "*.jpg*"),
                                                       ("all files",
                                                        "*.*")))
    head, tail = ntpath.split(filename)  
    ROOT_PATH = head
    print(ROOT_PATH)
      
    # Change label contents
    filenameStr.set(tail)
    
    
def predict(filename,predictStr):
    test_image = cv2.imread(os.path.join(ROOT_PATH,filename),cv2.IMREAD_COLOR)
    #print(test_image.shape)
    test_image = cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB)
    test_image = cv2.resize(test_image,(64,64), interpolation = cv2.INTER_AREA)
#     plt.imshow(test_image)
#     plt.show()
    #print(test_image.shape)
    test_image_expanded = np.expand_dims(test_image, axis = 0)
    #print(test_image.shape)
    test_image_expanded =test_image_expanded*1/255.0
    result = pretrained_model.predict_classes(test_image_expanded)
    
    predictStr.set("Predicted lable -- "+str(result[0]))

    # Driver code 
if __name__ == "__main__":
    
    # create a GUI window 
    gui = Tk() 
    gui.configure(background="#d6d6c2")
    gui.title("Classification GUI- Great Learning") 
    gui.geometry("680x300") 
    
    ### First Row ########
    filenamelabelStr = StringVar()
    
    filenamelabel = Label( gui, textvariable=filenamelabelStr,background="#d6d6c2",anchor="w")
    filenamelabelStr.set("Step 1: File Name")
    filenamelabel.grid(row=1, column=0,padx=20)
    
    filenameStr = StringVar()
    filenameValue = Entry(gui, textvariable=filenameStr) 
    filenameValue.grid(row=1, column=1,padx=20)
    
    importDataBtn = Button(gui, text='Import Data', fg='black', bg='#b3b3ff',padx="10", 
                    command=lambda:browseFiles(filenameStr), height=1, width=15,cursor = "hand2") 
    importDataBtn.grid(row=1, column=2,padx=20) 
    
    classifilabelStr = StringVar()
    classifilabelStr.set("Image classifier")
    classifilabel = Label( gui, textvariable=classifilabelStr,background="#d6d6c2",anchor="w")
    classifilabel.grid(row=2, column=0,padx=20)
    
    predictBtn = Button(gui, text='Predict', fg='black', bg='#b3b3ff',padx="10", 
                    command=lambda:predict(filenameStr.get(),predictStr), height=1, width=15,cursor = "hand2") 
    predictBtn.grid(row=3, column=1,padx=20) 
    
    predictStr = StringVar()
    predictlabel = Label( gui, textvariable=predictStr,background="#d6d6c2",anchor="w")    
    predictlabel.grid(row=3, column=2,padx=20)
    
    
                                                                                                  

    gui.mainloop()
```
