### Matrix Multipliction rule



 A - ( m1,n1 ) B - (m2,n2)

1. Matrix representation - ( rows, columns)
2. The number of columns in the first matrix must be equal to the number of rows in the second matrix. ( n1 must be equal to m2 )
3. The resulting matrix will have the same number of rows as the first matrix and the same number of columns as the second matrix --- resulting matrix ( m1,n2)
