```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([0,1,2,3,4])
y = np.array([1,2,3,4,5])

sns.lineplot(x=x,y=y,markers=True,dashes=False)
plt.plot(x,y,"o")
plt.xlim(0,5)
plt.ylim(0,5)
plt.show();
```

```python
x = np.array([0,1,2,3,4])
y = np.sin([0,1,2,3,4])
#sns.regplot(x,y)
plt.plot(x,y,scaley=True)
```
